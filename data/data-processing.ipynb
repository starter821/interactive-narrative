```python
import pandas as pd
import matplotlib.pyplot as plt

cpi_All = pd.read_csv('cpi_all_items.csv')
cpi_All['date'] = pd.to_datetime(cpi_All[['year', 'month']].assign(day=1)).dt.to_period('M')
cpi_All.head()
```

```python
base = cpi_All[(cpi_All['year'] == 2000) & (cpi_All['month'] == 2)]['cpi'].values[0]
cpi_All['pct_from_2000'] = (cpi_All['cpi'] - base) / base * 100
```

```python
cpi_All['pct_yoy'] = cpi_All['cpi'].pct_change(periods=6) * 100
cpi_All.head(10)
```

```python
cpi_All.to_csv('all_items_cpi_rebased.tsv', sep='\t', index=False)
```
